### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
print(tienda.head())
```

```python
print(tienda.columns)
```

```python
def facturacion(tienda):
  return tienda["Precio"].sum()

print("Facturación Tienda 1: $", facturacion(tienda))
print("Facturación Tienda 2: $", facturacion(tienda2))
print("Facturación Tienda 3: $", facturacion(tienda3))
print("Facturación Tienda 4: $", facturacion(tienda4))
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt

def ventas_categoria(tienda):
  return tienda["Categoría del Producto"].value_counts()

def grafica_ventas_categoria(tienda, nombre_tienda):
    conteo = tienda["Categoría del Producto"].value_counts()
    plt.figure(figsize=(10,5))
    plt.bar(conteo.index, conteo.values, color='skyblue')
    plt.title(f"Ventas por Categoría - {nombre_tienda}")
    plt.xlabel("Categoría")
    plt.ylabel("Cantidad de ventas")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
print("Ventas por categoría de Tienda 1:",ventas_categoria(tienda))
print("Ventas por categoría de Tienda 2:",ventas_categoria(tienda2))
print("Ventas por categoría de Tienda 3:",ventas_categoria(tienda3))
print("Ventas por categoría de Tienda 4:",ventas_categoria(tienda4))

#Gráfica de Los resultados:
grafica_ventas_categoria(tienda,"Tienda 1")

```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio(tienda, nombre_tienda):
    promedio = tienda["Calificación"].mean()
    print(f"Calificación promedio de {nombre_tienda}: {promedio:.2f}")
    return promedio
```

```python
calificacion_promedio(tienda, "Tienda 1")
calificacion_promedio(tienda2, "Tienda 2")
calificacion_promedio(tienda3, "Tienda 3")
calificacion_promedio(tienda4, "Tienda 4")
```

# 4. Productos más y menos vendidos

```python
def productos_masmenos_vendidos(tienda, nombre_tienda, top_n=10):
    conteo_productos = tienda["Producto"].value_counts()

    print(f"\nTop {top_n} productos más vendidos en {nombre_tienda}:\n")
    print(conteo_productos.head(top_n))

    print(f"\nTop {top_n} productos menos vendidos en {nombre_tienda}:\n")
    print(conteo_productos.tail(top_n))
```

```python
productos_masmenos_vendidos(tienda, "Tienda 1")
productos_masmenos_vendidos(tienda2, "Tienda 2")
productos_masmenos_vendidos(tienda3, "Tienda 3")
productos_masmenos_vendidos(tienda4, "Tienda 4")
```

```python
def grafico_pastel_productos(tienda, nombre_tienda):
    conteo = tienda["Producto"].value_counts().head(5)
    plt.figure(figsize=(6, 6))
    plt.pie(conteo, labels=conteo.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"Top 5 productos más vendidos - {nombre_tienda}")
    plt.show()

grafico_pastel_productos(tienda, "Tienda 1")

```

# 5. Envío promedio por tienda

```python
def costos_envio_promedio(tienda, nombre_tienda):
 promedio=tienda["Costo de envío"].mean()

 print(f"El costo de envío promedio de {nombre_tienda} es: ${promedio:.2f}")
```

```python
costos_envio_promedio(tienda, "Tienda 1")
costos_envio_promedio(tienda2, "Tienda 2")
costos_envio_promedio(tienda3, "Tienda 3")
costos_envio_promedio(tienda4, "Tienda 4")
```

```python
#Gráfico de dispersión para comparar precio vs. costo de envío
def grafico_dispersion_precio_envio(tienda, nombre_tienda):
    plt.figure(figsize=(8, 5))
    plt.scatter(tienda["Precio"], tienda["Costo de envío"], alpha=0.5)
    plt.title(f"Precio vs Costo de Envío - {nombre_tienda}")
    plt.xlabel("Precio")
    plt.ylabel("Costo de Envío")
    plt.grid(True)
    plt.show()

grafico_dispersion_precio_envio(tienda,"Tienda 1")
```
